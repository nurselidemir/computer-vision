#  OCR (Optik Karakter TanÄ±ma) Nedir ve NasÄ±l YapÄ±lÄ±r?

##  OCR Nedir?

**OCR (Optical Character Recognition)**, taranmÄ±ÅŸ belgelerdeki, gÃ¶rsellerdeki ya da el yazÄ±sÄ± metinlerdeki karakterleri tanÄ±yÄ±p dijital metne dÃ¶nÃ¼ÅŸtÃ¼ren bir teknolojidir.

Bu teknoloji sayesinde:
- GÃ¶rsellerden yazÄ±lar okunabilir,
- Faturalar, fiÅŸler gibi belgeler dijitalleÅŸtirilebilir,
- GÃ¶rme engellilere yÃ¶nelik uygulamalar geliÅŸtirilebilir,
- Arama motorlarÄ± gÃ¶rsellerdeki yazÄ±larÄ± tanÄ±yabilir.

## OCR NasÄ±l Ã‡alÄ±ÅŸÄ±r?

1. **GÃ¶rÃ¼ntÃ¼ Ã–n Ä°ÅŸleme (Preprocessing)**  
   - GÃ¼rÃ¼ltÃ¼ azaltma (Noise reduction)
   - Gri tonlamaya Ã§evirme
   - EÅŸikleme (Thresholding)
   - Kenar bulma (Edge detection)

2. **Karakter TanÄ±ma**  
   - Segmentasyon: Harfleri ayÄ±rma
   - SÄ±nÄ±flandÄ±rma: Her karakteri karÅŸÄ±lÄ±k gelen harf/sayÄ± ile eÅŸleÅŸtirme

3. **DoÄŸruluk ArtÄ±rma**  
   - Dil modeli yardÄ±mÄ±yla kelime tahmini
   - Optimizasyon algoritmalarÄ±

```python
from ultralytics import YOLO
import cv2
import os

model = YOLO('runs/detect/train19/weights/best.pt')
results = model.predict(source='dataset/images/val', imgsz=640)

save_dir = "runs/detect/cropped_plates"
os.makedirs(save_dir, exist_ok=True)

for i, result in enumerate(results):
    boxes = result.boxes.xyxy.cpu().numpy()  # kutu koordinatlarÄ±
    img_path = result.path
    img = cv2.imread(img_path)
    for j, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        cropped_plate = img[y1:y2, x1:x2]
        cv2.imwrite(f"{save_dir}/plate_{i}_{j}.jpg", cropped_plate)

print("Plakalar kÄ±rpÄ±ldÄ± ve kaydedildi.")

```

```python
import cv2
import pytesseract
import random
import os

pytesseract.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'  
folder = '/Users/nurselidemir/Desktop/computer_vision/runs/detect/cropped_plates'


all_files = [f for f in os.listdir(folder) if f.endswith('.jpg')]

sample_files = random.sample(all_files, min(20, len(all_files)))

for file in sample_files:
    img_path = os.path.join(folder, file)
    img = cv2.imread(img_path)

   
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray)  
    gray = cv2.medianBlur(gray, 3) 

    custom_config = r'--psm 7 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'

    text = pytesseract.image_to_string(gray, config=custom_config)

    text = text.strip()
    if len(text) == 0 or any(c not in "ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789" for c in text):
        text = "OKUNAMADI"

    print(f'Plaka: {file} --> {text}')
```
